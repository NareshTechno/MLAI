```python
import numpy as np
import pandas as pd
import json
from pathlib import Path


from sklearn import datasets, linear_model, metrics

folder = Path('C:/Upgrad Projects/LogisticRegression/diabetes_example')

# Reading the file
diabetes_train  = pd.read_csv(folder/'diabetes_train.csv')

diabetes_test  = pd.read_csv(folder/'diabetes_test.csv')

# read the file containing actual labels (for computing scores)
actual_data = pd.read_csv(folder/"sample_diabetes_actual.csv")


print(diabetes_train.head())
print(diabetes_test.head())
print(actual_data.head())

## WRITE YOUR CODE HERE ##

# create logistic regression object
log_model = linear_model.LogisticRegression()

#  create x_train and y_train
x_train = diabetes_train.iloc[:, :7]
y_train = diabetes_train.iloc[:, 7]


# Fit the model
log_model.fit(x_train, y_train)

# make predictions
predictions = log_model.predict(diabetes_test.iloc[:, :7])
y_pred = pd.DataFrame(predictions)

# Compute evaluation scores and round them off to two decimal places
# note that the test case may fail if the metrics are not rounded off 
accuracy_score = round(metrics.accuracy_score(actual_data.Diabetes, predictions),2)
precision_score = round(metrics.precision_score(actual_data.Diabetes, predictions),2)
recall_score = round(metrics.recall_score(actual_data.Diabetes, predictions),2)
f1_score = round(metrics.f1_score(actual_data.Diabetes, predictions),2)


print(accuracy_score, precision_score, recall_score, f1_score)


# write the scores into the output file as a dictionary
# this code is already written for you
###
d = {'recall_score': recall_score,
    'f1_score': f1_score,
    'accuracy_score': accuracy_score,
	'precision_score': precision_score}
print(d)

# writing the dictionary to output file
# code is already written
with open(folder/'answers.txt', 'w') as file:
     file.write(json.dumps(d)) 
```
